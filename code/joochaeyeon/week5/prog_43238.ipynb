```python
def solution(n, times):
    answer = 0 #  

    left, right = 1, max(times)*n #       

    while left <= right: #    <= 
        middle = (left+right)//2
        totalPeople = sum(middle//time for time in times)

        if totalPeople >=n: #     
            answer = middle
            right=middle-1
        else:
            left=middle+1

    return answer


# 
n = 6  #     n
times = [7, 10]  #       
print(solution(n, times))
```
